## 1. DESCRIPTION

Dans le cadre de ce projet, vous effectuerez une annotation de questions en langue naturelle en utilisant différents modèles avec pour objectif ultime de proposer une méthode qui dépasse les performances des équipes concurrentes.

Par exemple, étant donnée la question:
What is the country for head of state of Justin Trudeau ?
Votre modèle devra retourner :
what is the <\<wd:Q6256\>> for <\<wdt:P35\>> of <\<wd:Q3099714\>> ?

Les éléments ayant pour préfixe wd sont des URIs de Wikidata. Par exemple wd:Q6256 correspond à l'URI https://www.wikidata.org/wiki/Q6256

Wikidata est une base de données collaborative et structurée qui fait partie du projet Wikimedia, tout comme Wikipédia. Contrairement à Wikipédia, qui se concentre sur la création et la gestion de contenu encyclopédique, Wikidata se spécialise dans la collecte et la gestion de données structurées.

**Ressources Wikidata :**

Les ressources sont des entités du monde réel avec un identifiant unique précédé du préfixe "wd".

Voici quelques exemples d'éléments :

1. **La Tour Eiffel (Identifiant : Q243)** - Il s'agit d'un élément représentant la célèbre tour située à Paris.

2. **Albert Einstein (Identifiant : Q937)** - C'est un élément qui représente le physicien renommé.

3. **Mona Lisa (La Joconde) (Identifiant : Q12418)** - Un élément correspondant à l'œuvre d'art iconique de Leonardo da Vinci.

**Propriétés Wikidata :**

Les propriétés sont des caractéristiques ou des informations spécifiques associées aux éléments. Chaque propriété a un identifiant unique avec l'un des préfixes suivant: "wdt", "ps" ou "pq" et est utilisée pour décrire, quantifier ou qualifier les éléments. Voici des exemples de propriétés avec les éléments correspondants :

1. **Hauteur (Propriété : P2049)** - Une propriété utilisée pour indiquer la hauteur d'une structure. Par exemple, la hauteur de la Tour Eiffel est stockée dans cette propriété.

2. **Date de naissance (Propriété : P569)** - Une propriété pour enregistrer la date de naissance des individus. Par exemple, la date de naissance d'Albert Einstein est stockée grâce à cette propriété.

3. **Créateur (Propriété : P170)** - Cette propriété est utilisée pour spécifier qui a créé une œuvre d'art. Dans le cas de la Mona Lisa, on associe cette propriété à l'élément correspondant pour indiquer que Leonardo da Vinci en est le créateur.

```python
if torch.cuda.is_available():
    device = torch.device("cuda")
else:
    device = torch.device("cpu")
# initialize the tokenizer
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
```

# Training

### Encode and tokenize raw sentences in train df

```python
question_raw = train_df['question_raw']
```

```python
input_ids = []
attention_masks = []
# Add special tokens (PAD, CLS, SEP) the tokenized raw sentences
for sent in question_raw:
    encoded_dict = tokenizer.encode_plus(
                        sent,
                        add_special_tokens = True,
                        truncation = True,
                        max_length = 256,
                        pad_to_max_length = True,
                        return_attention_mask = True,
                        return_tensors = 'pt',
                   )
    input_ids.append(encoded_dict['input_ids'][0])
    attention_masks.append(encoded_dict['attention_mask'][0])
```

### Align labels and tokenized raw question

```python
question_original = train_df['question_original'].tolist()
# replace every instance of {{ with { and }} with } and {} with ''
question_original = [question.replace('{{', '{').replace('}}', '}').replace('{}', '') for question in question_original]
# split {x and x} into { x and x }
question_original = [question.replace('{', '{ ').replace('}', ' }').replace('{}', '') for question in question_original]
# split the question into a list of tokens
question_original = [question.split() for question in question_original]
# tokenize original and use the result to build labels for the raw sentences
question_original = [' '.join(question) for question in question_original]
question_original = [tokenizer.tokenize(question) for question in question_original]
```

```python
raw_tokens = [tokenizer.convert_ids_to_tokens(ids) for ids in input_ids]
train_df["raw_tokens"] = raw_tokens
# make new column by looping through question_tagged and taking only whats inside the << >> to retrieve the wikidata tags associated with each question
tags = []
for i in range(len(train_df)):
    tags.append(re.findall(r'<<([^>]*)>>', train_df["question_tagged"][i]))
# add tags to the dataframe
train_df["tags"] = tags
train_df.head()
```

```python
# create labels for raw_tokens using question_original and tags to add bio_tags to the dataframe
labels = []
for row in range(len(train_df)):
    original_tokens = question_original[row]
    tagged_entities = tags[row]
    bio_tags = []
    tag_idx = 0

    inside_entity = False
    is_first_token = True
    for token in original_tokens:
        # skip special tokens or tokens that arent part of the bert vocabulary
        if token == "[CLS]" or token == "[SEP]" or token == "[PAD]" or token[0:2] == '##':
            continue
        if token == "{":
            inside_entity = True
            is_first_token = True
        elif token == "}":
            inside_entity = False
            tag_idx += 1
        # create B tags
        elif inside_entity and is_first_token:
            bio_tags.append("B-" + tagged_entities[tag_idx])
            is_first_token = False
        # create I tags
        elif inside_entity:
            bio_tags.append("I-" + tagged_entities[tag_idx])
        # if not inside an entity, assign O tag
        else:
            bio_tags.append("O")
    labels.append(bio_tags)
# add labels to the dataframe
train_df["labels"] = labels
train_df.head()
```

```python
print('Original Question: ', question_original[0])
print('Raw Question: ', question_raw[0])
print('Tokens from input Ids: ',  tokenizer.convert_ids_to_tokens(input_ids[0]))
print('BIO Labels:' , labels[0])
```

##### Now that we have a BIO label associated to each of the tokens from the tokenized raw sentences, we want to build a dictionary that relates the tag to its index so we can pass it to the model as a numerical representation

```python
tag2idx = {}
idx = 0
for tags in labels:
    for tag in tags:
        if tag not in tag2idx:
            tag2idx[tag] = idx
            idx += 1

# save tag2idx to a file so we can use it later
with open('data/tag2idx.json', 'w') as fp:
    json.dump(tag2idx, fp)
```

### Assign -100 to represent the special tokens and subwords that arent part of the BERT vocabulary

```python
new_labels = []
null_label_id = -100

for (inputs, labels_list) in zip(input_ids, labels):
    padded_labels = []
    idx = 0

    for token_id in inputs:
        token_id = token_id.numpy().item()
        # Assign -100 to the token if it is a special token or if it is a word that isnt part of the bert vocabulary
        if (token_id == tokenizer.pad_token_id) or \
            (token_id == tokenizer.cls_token_id) or \
            (token_id == tokenizer.sep_token_id) or \
            (tokenizer.ids_to_tokens[token_id][0:2]) == '##':
            padded_labels.append(null_label_id)
        else:
            # Assign the corresponding index from the dictionary and go to the next label
            label = labels_list[idx]
            padded_labels.append(tag2idx[label])
            idx += 1
    new_labels.append(padded_labels)
```

```python
# Convert the lists into PyTorch tensors.
pt_input_ids = torch.stack(input_ids, dim=0)
pt_attention_masks = torch.stack(attention_masks, dim=0)
pt_labels = torch.tensor(new_labels, dtype=torch.long)
train_dataset = TensorDataset(pt_input_ids, pt_attention_masks, pt_labels)
batch_size = 16
train_dataloader = DataLoader(train_dataset, sampler = RandomSampler(train_dataset), batch_size = batch_size)
```

```python
model = BertForTokenClassification.from_pretrained(
    "bert-base-uncased",
    num_labels = len(tag2idx),
    output_attentions = False,
    output_hidden_states = False,
)
model.to(device)
```

```python
# Load the AdamW optimizer
optimizer = AdamW(model.parameters(), lr = 5e-5, eps = 1e-8)
```

```python
epochs = 8
total_steps = len(train_dataloader) * epochs
# Create the learning rate scheduler.
scheduler = get_linear_schedule_with_warmup(optimizer,
                                            num_warmup_steps = 0,
                                            num_training_steps = total_steps)
```

```python
def format_time(elapsed):
    '''
    Takes a time in seconds and returns a string hh:mm:ss
    '''
    # Round to the nearest second.
    elapsed_rounded = int(round((elapsed)))

    # Format as hh:mm:ss
    return str(datetime.timedelta(seconds=elapsed_rounded))
```

### Training loop, adapted from the inf8460_BERT_Fine_Tuning_NER notebook provided in class

```python
# if model already saved as ./bert_trained_model, then skip this cell
if (os.path.exists('./bert_trained_model')):
    print('BERT model already exists as bert_trained_model, skipping training')
else:
    # This training code is based on the `run_glue.py` script here:
    # https://github.com/huggingface/transformers/blob/5bfcd0485ece086ebcbed2d008813037968a9e58/examples/run_glue.py#L128

    # Set the seed value all over the place to make this reproducible.
    seed_val = 42

    random.seed(seed_val)
    np.random.seed(seed_val)
    torch.manual_seed(seed_val)
    torch.cuda.manual_seed_all(seed_val)

    # Store the average loss after each epoch so we can plot them.
    loss_values = []

    # For each epoch...
    for epoch_i in range(0, epochs):

        # ========================================
        #               Training
        # ========================================

        # Perform one full pass over the training set.

        print("")
        print('======== Epoch {:} / {:} ========'.format(epoch_i + 1, epochs))
        print('Training...')

        # Measure how long the training epoch takes.
        t0 = time.time()

        # Reset the total loss for this epoch.
        total_loss = 0

        # Put the model into training mode. Don't be mislead--the call to
        # `train` just changes the *mode*, it doesn't *perform* the training.
        # `dropout` and `batchnorm` layers behave differently during training
        # vs. test (source: https://stackoverflow.com/questions/51433378/what-does-model-train-do-in-pytorch)
        model.train()

        # For each batch of training data...
        for step, batch in enumerate(train_dataloader):

            # Progress update every 40 batches.
            if step % 40 == 0 and not step == 0:
                # Calculate elapsed time in minutes.
                elapsed = format_time(time.time() - t0)

                # Report progress.
                print('  Batch {:>5,}  of  {:>5,}.    Elapsed: {:}.'.format(step, len(train_dataloader), elapsed))

            # Unpack this training batch from our dataloader.
            #
            # As we unpack the batch, we'll also copy each tensor to the GPU using the
            # `to` method.
            #
            # `batch` contains three pytorch tensors:
            #   [0]: input ids
            #   [1]: attention masks
            #   [2]: labels
            b_input_ids = batch[0].to(device)
            b_input_mask = batch[1].to(device)
            b_labels = batch[2].to(device)
            # b_input_ids = batch[0]
            # b_input_mask = batch[1]
            # b_labels = batch[2]

            # Always clear any previously calculated gradients before performing a
            # backward pass. PyTorch doesn't do this automatically because
            # accumulating the gradients is "convenient while training RNNs".
            # (source: https://stackoverflow.com/questions/48001598/why-do-we-need-to-call-zero-grad-in-pytorch)
            model.zero_grad()

            # In PyTorch, calling `model` will in turn call the model's `forward`
            # function and pass down the arguments. The `forward` function is
            # documented here:
            # https://huggingface.co/transformers/model_doc/bert.html#bertfortokenclassification
            # The results are returned in a results object, documented here:
            # https://huggingface.co/transformers/main_classes/output.html#transformers.modeling_outputs.TokenClassifierOutput
            result = model(b_input_ids,
                        token_type_ids=None,
                        attention_mask=b_input_mask,
                        labels=b_labels)

            loss = result.loss

            # Accumulate the training loss over all of the batches so that we can
            # calculate the average loss at the end. `loss` is a Tensor containing a
            # single value; the `.item()` function just returns the Python value
            # from the tensor.
            total_loss += loss.item()

            # Perform a backward pass to calculate the gradients.
            loss.backward()

            # Clip the norm of the gradients to 1.0.
            # This is to help prevent the "exploding gradients" problem.
            torch.nn.utils.clip_grad_norm_(model.parameters(), 1.0)

            # Update parameters and take a step using the computed gradient.
            # The optimizer dictates the "update rule"--how the parameters are
            # modified based on their gradients, the learning rate, etc.
            optimizer.step()

            # Update the learning rate.
            scheduler.step()

        # Calculate the average loss over the training data.
        avg_train_loss = total_loss / len(train_dataloader)

        # Store the loss value for plotting the learning curve.
        loss_values.append(avg_train_loss)

        print("")
        print("  Average training loss: {0:.2f}".format(avg_train_loss))
        print("  Training epcoh took: {:}".format(format_time(time.time() - t0)))


    print("")
    print("Training complete!")
```

```python
def create_wikidata_T5_tokenizer(tokenizer_model_name = 't5-base', fast = False):

    tokenizer = AutoTokenizer.from_pretrained(tokenizer_model_name, use_legacy=False, use_fast=fast)
    new_token = tokenizers.AddedToken('<<', lstrip=False, rstrip=False, normalized=False, single_word=False)
    tokenizer.add_tokens(new_token, special_tokens=False)
    #tokenizer.add_tokens('<<', special_tokens=False)
    return tokenizer

tokenizer = create_wikidata_T5_tokenizer()

sentence = ' this is a test <<wdt:P22>> sentence'

tokenized_sentence = tokenizer.encode(sentence)
print('Sentence :', sentence)
print('Encoded  : ', tokenized_sentence)
print('Decoded  :', tokenizer.decode(tokenized_sentence))
```

```python
from torch.utils.data import Dataset, DataLoader
import tensorflow as tf

class WikidataDataset(Dataset):
  def __init__(self, tokenizer, dataset, type_path, max_len=MAX_LEN):

    self.data = dataset[type_path] # Select between train, val & test set
    self.max_len = max_len
    self.tokenizer = tokenizer
    self.inputs = []
    self.targets = []

    self._build()

  def __len__(self):
    return len(self.inputs)

  def __getitem__(self, index):
    source_ids = self.inputs[index]["input_ids"].squeeze()
    src_mask    = self.inputs[index]["attention_mask"].squeeze()

    if self.targets != []:
      target_ids = self.targets[index]["input_ids"].squeeze()
      target_mask = self.targets[index]["attention_mask"].squeeze()
      return {"source_ids": source_ids, "source_mask": src_mask, "target_ids": target_ids, "target_mask": target_mask}
    else:
      return {"source_ids": source_ids, "source_mask": src_mask, "target_ids": torch.zeros(self.max_len), "target_mask": torch.zeros(self.max_len)}

  def _build(self):
    for idx in range(len(self.data)):
      # tokenize the sentence inputs

      input_ = self.data[idx]["question_raw"]
      target = self.data[idx]["question_tagged"]

      tokenized_inputs = self.tokenizer.batch_encode_plus(
          [input_], max_length=self.max_len, padding="max_length", truncation=False, return_tensors="pt"
      )

      self.inputs.append(tokenized_inputs)

      if target is not None: # Test set only have inputs

        # tokenize targets
        tokenized_targets = self.tokenizer.batch_encode_plus(
            [target],max_length=self.max_len, padding="max_length", truncation=False, return_tensors="pt"
        )

        self.targets.append(tokenized_targets)
```

```python
def get_dataset(tokenizer, type_path, args):
    tokenizer.max_length = args.max_seq_length
    tokenizer.model_max_length = args.max_seq_length
    return WikidataDataset(tokenizer=tokenizer, dataset=dataset, type_path=type_path, max_len=args.max_seq_length)
```

##### Model

Majority of the code here is adapted from [here](https://colab.research.google.com/github/patil-suraj/exploring-T5/blob/master/t5_fine_tuning.ipynb) which uses the pytorch-lightning framework for training neural networks. T5 has shown that it can generate state of the art on many tasks as long as it can be cast as a text-to-text problem

```python
class T5FineTuner(pl.LightningModule):
    def __init__(self, hparam):
        super(T5FineTuner, self).__init__()
        self.hparam = hparam

        self.model = T5ForConditionalGeneration.from_pretrained(hparam.model_name_or_path)
        self.tokenizer = create_wikidata_T5_tokenizer(hparam.tokenizer_name_or_path)

        #self.model.resize_token_embeddings(len(self.tokenizer))
        self.save_hyperparameters()

        self.train_outputs =  []
        self.val_outputs =  []

        # https://stackoverflow.com/questions/73111496/pytorch-lightning-misconfiguration-exception-the-closure-hasnt-been-executed
        self.automatic_optimization = False

    def is_logger(self):
        return True

    def forward(
        self, input_ids, attention_mask=None, decoder_input_ids=None, decoder_attention_mask=None, lm_labels=None
    ):
        return self.model(
            input_ids,
            attention_mask=attention_mask,
            decoder_input_ids=decoder_input_ids,
            decoder_attention_mask=decoder_attention_mask,
            labels=lm_labels,
        )

    def _step(self, batch):
        lm_labels = batch["target_ids"]
        lm_labels[lm_labels[:, :] == self.tokenizer.pad_token_id] = -100

        outputs = self(
            input_ids=batch["source_ids"],
            attention_mask=batch["source_mask"],
            lm_labels=lm_labels,
            decoder_attention_mask=batch['target_mask']
        )

        loss = outputs[0]

        return loss

    def training_step(self, batch, batch_idx):

        loss = self._step(batch)
        tensorboard_logs = {"train_loss": loss}

        train_output = {"loss": loss, "log": tensorboard_logs}
        self.train_outputs.append(train_output)

        self.log("my_loss", loss, on_step=True, on_epoch=True, prog_bar=True, logger=True)

        # https://stackoverflow.com/questions/73111496/pytorch-lightning-misconfiguration-exception-the-closure-hasnt-been-executed
        # Insert these lines:
        self.manual_backward(loss)

        optimizer = self.optimizers()
        scheduler = self.lr_scheduler

        optimizer.step()
        optimizer.zero_grad()
        scheduler.step()

        return train_output

    def on_train_epoch_end(self):
        avg_train_loss = torch.stack([x["loss"] for x in self.train_outputs]).mean()
        tensorboard_logs = {"avg_train_loss": avg_train_loss}

        self.train_outputs = []

    def validation_step(self, batch, batch_idx):
        loss = self._step(batch)

        val_output = {"val_loss": loss}
        self.val_outputs.append(val_output)

        self.log("my_loss_val", loss, on_step=True, on_epoch=True, prog_bar=True, logger=True)

        return val_output

    def on_validation_epoch_end(self):
        avg_loss = torch.stack([x["val_loss"] for x in self.val_outputs]).mean()
        tensorboard_logs = {"val_loss": avg_loss}

        self.val_outputs = []

    def configure_optimizers(self):
        "Prepare optimizer and schedule (linear warmup and decay)"

        model = self.model
        no_decay = ["bias", "LayerNorm.weight"]
        optimizer_grouped_parameters = [
            {
                "params": [p for n, p in model.named_parameters() if not any(nd in n for nd in no_decay)],
                "weight_decay": self.hparam.weight_decay,
            },
            {
                "params": [p for n, p in model.named_parameters() if any(nd in n for nd in no_decay)],
                "weight_decay": 0.0,
            },
        ]
        optimizer = AdamW(optimizer_grouped_parameters,
                          lr=self.hparam.learning_rate, eps=self.hparam.adam_epsilon)
        self.opt = optimizer
        return [optimizer]

    def optimizer_step(self,
                       epoch=None,
                       batch_idx=None,
                       optimizer=None,
                       optimizer_idx=None,
                       optimizer_closure=None,
                       on_tpu=None,
                       using_native_amp=None,
                       using_lbfgs=None
                       ):

        optimizer.step(closure=optimizer_closure)
        optimizer.zero_grad()
        self.lr_scheduler.step()

    def get_tqdm_dict(self):
        tqdm_dict = {"loss": "{:.3f}".format(
            self.trainer.avg_loss), "lr": self.lr_scheduler.get_last_lr()[-1]}

        return tqdm_dict

    def train_dataloader(self):
        train_dataset = get_dataset(
            tokenizer=self.tokenizer, type_path="train", args=self.hparam)
        dataloader = DataLoader(train_dataset, batch_size=self.hparam.train_batch_size,
                                drop_last=True, shuffle=True, num_workers=2)
        t_total = (
            (len(dataloader.dataset) //
             (self.hparam.train_batch_size * max(1, self.hparam.n_gpu)))
            // self.hparam.gradient_accumulation_steps
            * float(self.hparam.num_train_epochs)
        )
        scheduler = get_linear_schedule_with_warmup(
            self.opt, num_warmup_steps=self.hparam.warmup_steps, num_training_steps=t_total
        )
        self.lr_scheduler = scheduler
        return dataloader

    def val_dataloader(self):
        val_dataset = get_dataset(
            tokenizer=self.tokenizer, type_path="validation", args=self.hparam)
        return DataLoader(val_dataset, batch_size=self.hparam.eval_batch_size, num_workers=2)
```

```python
logger = logging.getLogger(__name__)

class LoggingCallback(pl.Callback):
  def on_validation_end(self, trainer, pl_module):
    logger.info("***** Validation results *****")
    if pl_module.is_logger():
      metrics = trainer.callback_metrics
      # Log results
      for key in sorted(metrics):
        if key not in ["log", "progress_bar"]:
          logger.info("{} = {}\n".format(key, str(metrics[key])))

  def on_test_end(self, trainer, pl_module):
    logger.info("***** Test results *****")

    if pl_module.is_logger():
      metrics = trainer.callback_metrics

      # Log and save results to file
      output_test_results_file = os.path.join(pl_module.hparams.output_dir, "test_results.txt")
      with open(output_test_results_file, "w") as writer:
        for key in sorted(metrics):
          if key not in ["log", "progress_bar"]:
            logger.info("{} = {}\n".format(key, str(metrics[key])))
            writer.write("{} = {}\n".format(key, str(metrics[key])))
```

```python
def generate_outputs(model, dataset=dataset, type_path=None):

  data = WikidataDataset(tokenizer=model.tokenizer, dataset=dataset, type_path=type_path)
  dataloader = DataLoader(data, batch_size=32, num_workers=2, shuffle=False)

  if torch.cuda.is_available():
      device = torch.device("cuda")
      print("Using GPU:", torch.cuda.get_device_name(0))
  else:
      device = torch.device("cpu")
      print("Using CPU")

  model.model.eval()
  model = model.to(device)

  outputs = []
  targets = []
  texts = []
  for batch in tqdm(dataloader, position=0, leave=0):

    if torch.cuda.is_available():
      input_ids = batch['source_ids'].to("cuda")
      attention_mask = batch['source_mask'].to("cuda")
      if type_path != 'test':
        target_ids = batch["target_ids"].to("cuda")
    else:
      input_ids = batch['source_ids']
      attention_mask = batch['source_mask']
      if type_path != 'test':
        target_ids = batch["target_ids"]

    outs = model.model.generate(input_ids=input_ids,
                                attention_mask=attention_mask, min_length=10, max_length=args.max_seq_length)

    dec = [tokenizer.decode(ids, skip_special_tokens=True, clean_up_tokenization_spaces=False).strip() for ids in outs]
    outputs.extend(dec)

    if type_path != 'test':
      target = [tokenizer.decode(ids, skip_special_tokens=True, clean_up_tokenization_spaces=False).strip()
                  for ids in target_ids]
      targets.extend(target)

    text = [tokenizer.decode(ids, skip_special_tokens=True, clean_up_tokenization_spaces=False).strip()
                for ids in input_ids]
    texts.extend(text)

  return texts, outputs, targets

def save_generated_outputs(texts, outputs, targets, type_path=None, model_version=None):

  output_tags = find_and_save_tags(outputs, type_path, 'outputs', model_version)

  if targets is not None: # None for test dataset
    target_tags = find_and_save_tags(targets, type_path, 'targets', model_version)

  data = save_dataset_text_target_output(type_path, texts, outputs, targets, model_version)
```

```python
class LookupKB:
    def __init__(self, model_path="/content/drive/My Drive/models/model_kb"):
        self.nlp = spacy.load("en_core_web_sm")

        if os.path.exists(model_path):
            self.kb = InMemoryLookupKB(vocab=self.nlp.vocab, entity_vector_length=96)
            self.kb.from_disk(model_path)
            return

    def process_data_for_kb(self, question, tagged):
        question_matches = re.finditer(r"{([^}]+)}", question)
        tag_matches = re.finditer(r"<<([^>]+)>>", tagged)

        entity_data = []

        for q_match, t_match in zip(question_matches, tag_matches):
            alias = q_match.group(1)
            entity_id = t_match.group(1)
            entity_data.append((entity_id, alias, question))

        return entity_data

    def load_data(self, filename, output_file="/content/drive/My Drive/models/model_kb"):
        train_data = pd.read_csv(filename, delimiter="|")

        if os.path.exists(output_file):
            self.kb = InMemoryLookupKB(vocab=self.nlp.vocab, entity_vector_length=96)
            self.kb.from_disk(output_file)
            return
        else:
            self.kb = InMemoryLookupKB(vocab=self.nlp.vocab, entity_vector_length=96)

        added_entities = set()
        added_aliases = set()

        for _, row in tqdm.tqdm(train_data.iterrows(), total=len(train_data)):
            entity_data = self.process_data_for_kb(
                row["question_original"], row["question_tagged"]
            )
            for entity_id, alias, desc in entity_data:
                if entity_id not in added_entities:
                    desc_doc = self.nlp(desc)
                    desc_vector = desc_doc.vector
                    self.kb.add_entity(
                        entity=entity_id, entity_vector=desc_vector, freq=1
                    )
                    added_entities.add(entity_id)
                if alias not in added_aliases:
                    self.kb.add_alias(alias, entities=[entity_id], probabilities=[0.1])
                    added_aliases.add(alias)
        self.kb.to_disk(output_file)

    def get_candidates(self, text):
        doc = self.nlp(text)
        return self.kb.get_candidates(doc)

```

```python
class Predictor:
    def __init__(self, lkb, ner):
        self.lkb = lkb
        self.ner = ner

    def find_with_spaces(self, pattern, text):
        pattern = pattern.replace(" ", "")
        pattern_re = re.compile(" *".join(map(re.escape, pattern)))

        m = pattern_re.search(text)
        if m:
            return m.start(), m.end()

    def apply_labels_to_text(self, row, id_to_label):
        text = row["question_raw"]
        labels = id_to_label[id_to_label["question_id"] == row["question_id"]]["label"]

        updated_text = ""
        remaining_text = text

        for label in labels:
            label = label.replace("{", "")
            label_location = self.find_with_spaces(label.lower(), remaining_text)

            if label_location:
                label_start, label_end = label_location
                updated_text += remaining_text[:label_start] + "{" + label + "}"
                remaining_text = remaining_text[label_end:]

        updated_text += remaining_text

        return updated_text

    def align_predictions(self, row, id_to_label):
        question_id = row["question_id"]
        predicted_entities = row["predicted_entities"]
        target_labels = id_to_label[id_to_label["question_id"] == question_id][
            "label"
        ].tolist()
        aligned_predictions = []

        for label in target_labels:
            label = label.replace("{", "")
            prediction = next(
                (
                    entity
                    for entity in predicted_entities
                    if entity[0] == "{" + label + "}"
                ),
                ("", ""),
            )

            candidates_kb = [c.entity_ for c in self.lkb.get_candidates(label)]

            if not prediction[1] and candidates_kb:
                prediction = (prediction[0], candidates_kb[0])

            aligned_predictions.append(prediction[1])

        return aligned_predictions

    def get_id_to_label_row(self, question, tagged, question_id):
        """Process a single row of data."""
        question_matches = list(re.finditer(r"[{<]([^}>]+)[}>]", question))
        tag_matches = list(re.finditer(r"<<([^>]+)>>", tagged))

        row_id_labels = []
        for q_match, t_match in zip(question_matches, tag_matches):
            label = q_match.group(1)
            wiki_id = t_match.group(1)

            row_id_labels.append(
                {
                    "question_id": question_id,
                    "label": label,
                    "wiki_id": wiki_id,
                }
            )

        return row_id_labels

    def get_id_to_label(self, df):
        """Process dataframe for classification."""
        processed_data = []

        for _, row in df.iterrows():
            processed_data.extend(
                self.get_id_to_label_row(
                    row["question_original"],
                    row["question_tagged"],
                    row["question_id"],
                )
            )

        return pd.DataFrame(processed_data)

    def evaluate(self, eval_file):
        eval_df = pd.read_csv(eval_file, delimiter="|")

        actuals_df = self.get_id_to_label(eval_df)
        predictions_df = self.get_predictions(eval_df, actuals_df)

        f1 = f1_score(
            actuals_df["wiki_id"], predictions_df["target"], average="weighted"
        )

        accuracy = accuracy_score(actuals_df["wiki_id"], predictions_df["target"])

        return f1, accuracy

    def get_predictions(self, predictions_df, id_to_label_df):
        predicted_entities = []
        for text in tqdm.tqdm(
            predictions_df["question_original"], desc="Predicting entities"
        ):
            predicted_entities.append(self.ner.predict(text))

        predictions_df["predicted_entities"] = predicted_entities

        predictions_df["aligned_predictions"] = predictions_df.apply(
            lambda row: self.align_predictions(row, id_to_label_df), axis=1
        )

        final_output = []
        id = 1
        for _, row in predictions_df.iterrows():
            predictions = row["aligned_predictions"]
            for pred in predictions:
                final_output.append({"Id": id, "target": pred})
                id += 1

        return pd.DataFrame(final_output)

    def generate_submission(self, test_file, id_to_label_file):
        test_df = pd.read_csv(test_file, delimiter="|")
        id_to_label_df = pd.read_csv(id_to_label_file)

        test_df["question_original"] = test_df.apply(
            lambda row: self.apply_labels_to_text(row, id_to_label_df), axis=1
        )

        final_df = self.get_predictions(test_df, id_to_label_df)

        return final_df
```
